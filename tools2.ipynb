```python
prompt = f"""
You are a book recommendation assistant.
You have access to a list of books with their details including price,
rating, description, and features.
Based on the user's preferences and requirements,
recommend the most suitable books from the list.
Available Books:
{products}
"""
```

```python
print(model.invoke(prompt).content)
```

```python
user_input = input("enter question")
```

```python
@tool
def get_book(user_input: str) -> str:
    """Recommend a book based on description"""

    for product in products:
        for book_name, details in product.items():

            if user_input.lower() in details["description"].lower():

                return f"""
Book: {book_name}
Price: {details['price']}
Rating: {details['rating']}
Description: {details['description']}
"""

    return "Book not found"
```

```python
agent = create_agent(
    model= model,
    tools=[get_book],
    system_prompt="you are a good assissestient recommend me a book  based on my content  "
        
)
```

```python
response = agent.invoke({
        "messages": [
            {
                "role": "user",
                "content": user_input
            }
        ]
    }
)

print(response)
```
