```python
def mul(*num):
    m = 1
    for i in num:
        m *= i
    return m
```

# scope of variables

## local scope

```python
count = 0   # global variable
def counter():
    global count
    #count=0   #local variable
    count += 1
    print(f"count inside function: {count}")
    
counter()
print(f"count outside of function: {count}")
```

```python
count = 0 
def outer():
    count = 0
    def inner():
        count = 0
        count += 1
        print(f"inner: {count}")
        
    inner()
    print(f"outer: {count}")

outer()

print(f"global: {count}")
```
